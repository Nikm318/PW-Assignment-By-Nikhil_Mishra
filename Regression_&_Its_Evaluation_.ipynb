Question 1:  What is Simple Linear Regression?

Answer:

Simple Linear Regression is a statistical method used to measure and predict the relationship between two variables:

One Independent Variable (X) ‚Üí the predictor

One Dependent Variable (Y) ‚Üí the outcome you want to predict

It assumes that the relationship between X and Y can be represented by a straight line:

Equation

Y=a+bX

Where:

a = intercept (value of Y when X = 0)

b = slope (how much Y changes for each 1-unit change in X)

Purpose

To understand if X influences Y

To predict Y for any given value of X

Example

If you want to predict house price (Y) based on house size (X), you can use simple linear regression.

In simple terms

It draws the best-fitting straight line through your data points to explain and predict outcomes.


Question 2: What are the key assumptions of Simple Linear Regression?

Answer:   
The key assumptions of Simple Linear Regression, explained clearly and simply:

1. Linearity

The relationship between X and Y must be linear.

Meaning:

a straight line should describe how Y changes with X.

2. Independence of Errors

Residuals (errors) should be independent of each other.

No pattern, no correlation between errors.

Example violation:

Time-series data where today's error affects tomorrow‚Äôs error.

3. Homoscedasticity (Constant Variance)

The variance of errors should be constant across all levels of X.

Residuals should not show increasing or decreasing spread.

If variance changes ‚Üí heteroscedasticity (bad for regression).

4. Normality of Errors

Residuals (not X or Y) should be normally distributed.

Important for hypothesis tests and confidence intervals.

5. No Perfect Multicollinearity

For simple linear regression, this is trivial because there is only one X.

But in general - predictors should not be perfectly correlated.

Question 3: What is heteroscedasticity, and why is it important to address in regression models?


Answer:

Heteroscedasticity occurs when the variance of the error terms (residuals) is not constant across all levels of the independent variable.


In simple words:

Residuals spread out unevenly ‚Äî sometimes wide, sometimes narrow ‚Äî instead of being consistently scattered.

 What It Looks Like

If you plot residuals vs. predicted values:

Homoscedasticity ‚Üí uniform scatter

Heteroscedasticity ‚Üí funnel shape, cone shape, or any pattern where spread increases/decreases.

 Why Is It Important to Address?
1. It makes coefficient estimates inefficient

OLS regression assumes constant variance.
When this is violated, the model still gives unbiased coefficients but they are not optimal (not minimum variance).

2. Standard errors become incorrect

Because standard errors rely on constant variance, heteroscedasticity causes them to be:

too high

or too low

This leads to:

3. Hypothesis tests become unreliable

t-tests

F-tests

Confidence intervals

All may give wrong conclusions - you may think a variable is significant when it's not (or vice versa).

4. Impacts prediction accuracy

Uneven variance leads to less reliable predictions, especially in regions where variance is high.

Question 4: What is Multiple Linear Regression?

Answer:  

Multiple Linear Regression (MLR) is an extension of simple linear regression that uses two or more independent variables (X‚ÇÅ, X‚ÇÇ, X‚ÇÉ, ‚Ä¶) to predict a single dependent variable (Y).

 Definition

Multiple Linear Regression is a statistical technique that models the relationship between a dependent variable and multiple predictors, assuming the relationship is linear.

Equation Y=a+b1‚ÄãX1‚Äã+b2‚ÄãX2‚Äã+b3‚ÄãX3‚Äã+‚ãØ+bn‚ÄãXn‚Äã

a ‚Üí Intercept

b‚ÇÅ, b‚ÇÇ, ‚Ä¶, b‚Çô ‚Üí Coefficients showing how much Y changes with each X, keeping others constant

 Purpose of MLR

To understand how multiple factors simultaneously affect an outcome

To predict the value of Y using several inputs

To determine which variables are significant predictors

 Example

Predicting house price (Y) using:

X‚ÇÅ = Size of house

X‚ÇÇ = Number of bedrooms

X‚ÇÉ = Location rating

Each predictor contributes to explaining the variation in Y, and the regression line becomes a hyperplane instead of a simple straight line.

Question 5: What is polynomial regression, and how does it differ from linear regression?

General form:

| When used    | When relationship between X and Y is linear | When relationship is **non-linear** but smooth   |
| Flexibility  | Low                                         | High (can capture curves, bends, U-shapes, etc.) |

```python
#Question 6:  Implement a Python program to fit a Simple Linear Regression model to the following sample data: ‚óè X = [1, 2, 3, 4, 5] ‚óè Y = [2.1, 4.3, 6.1, 7.9, 10.2] Plot the regression line over the data points.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Given Data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create and train the model
model = LinearRegression()
model.fit(X, Y)

# Predictions
Y_pred = model.predict(X)

# Print slope and intercept
print("Slope (b1):", model.coef_[0])
print("Intercept (a):", model.intercept_)

# Plotting
plt.scatter(X, Y, label='Data Points')
plt.plot(X, Y_pred, label='Regression Line')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression")
plt.legend()
plt.show()
```

```python
#Question 7: Fit a Multiple Linear Regression model on this sample data: ‚óè Area = [1200, 1500, 1800, 2000] ‚óè Rooms = [2, 3, 3, 4] ‚óè Price = [250000, 300000, 320000, 370000] Check for multicollinearity using VIF and report the results.
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm

# Sample Data
area = [1200, 1500, 1800, 2000]
rooms = [2, 3, 3, 4]
price = [250000, 300000, 320000, 370000]

# Create DataFrame
df = pd.DataFrame({
    "Area": area,
    "Rooms": rooms,
    "Price": price
})

# Independent variables (X) and dependent variable (Y)
X = df[["Area", "Rooms"]]
Y = df["Price"]

# Fit the Multiple Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Print coefficients
print("Intercept:", model.intercept_)
print("Coefficients:")
for col, coef in zip(X.columns, model.coef_):
    print(f"{col}: {coef}")

# -------------------------------
# VIF Calculation
# -------------------------------
# Add constant for statsmodels
X_with_const = sm.add_constant(X)

vif_data = pd.DataFrame()
vif_data["Feature"] = X_with_const.columns
vif_data["VIF"] = [variance_inflation_factor(X_with_const.values, i)
                   for i in range(X_with_const.shape[1])]

print("\nVIF Results:")
print(vif_data)

```

```python
#Question 8: Implement polynomial regression on the following data: ‚óè X = [1, 2, 3, 4, 5]‚óè Y = [2.2, 4.8, 7.5, 11.2, 14.7] Fit a 2nd-degree polynomial and plot the resulting curve.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Given Data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Create polynomial features (degree = 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit Polynomial Regression Model
model = LinearRegression()
model.fit(X_poly, Y)

# Predict values
X_range = np.linspace(1, 5, 100).reshape(-1, 1)
X_range_poly = poly.transform(X_range)
Y_pred = model.predict(X_range_poly)

# Print coefficients
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)

# Plotting
plt.scatter(X, Y, label="Data Points")
plt.plot(X_range, Y_pred, label="Polynomial Curve (Degree 2)")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("2nd-Degree Polynomial Regression")
plt.legend()
plt.show()
```

```python
#Question 9: Create a residuals plot for a regression model trained on this data: ‚óè X = [10, 20, 30, 40, 50] ‚óè Y = [15, 35, 40, 50, 65] Assess heteroscedasticity by examining the spread of residuals.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Given Data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Fit Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Predictions
Y_pred = model.predict(X)

# Residuals
residuals = Y - Y_pred

# Print residuals
print("Residuals:", residuals)

# Residual Plot
plt.scatter(X, residuals, label="Residuals")
plt.axhline(y=0, color='black', linestyle='--')  # reference line
plt.xlabel("X")
plt.ylabel("Residuals")
plt.title("Residual Plot to Check Heteroscedasticity")
plt.legend()
plt.show()

#If residuals are evenly scattered around 0 ‚Üí Homoscedasticity (good)

#If residuals form a pattern (funnel, cone, increasing spread) ‚Üí Heteroscedasticity (problematic)

#This plot does NOT show heteroscedasticity.
```

1. Addressing Heteroscedasticity

 Step 4: Consider Tree-Based or Nonlinear Models

Models like:

 2. Addressing Multicollinearity

Multicollinearity means predictors are strongly correlated, inflating variance of coefficients.

 Step 1: Diagnose Using VIF

Remove or adjust variables with VIF > 5 or 10

 Step 2: Remove Redundant Features

For example:

Area and number of rooms may be correlated

Bedroom size and number of bathrooms may correlate

Model stability on test data
