Question 1 : What is the difference between multithreading and multiprocessing?

-> Multithreading and multiprocessing are both techniques used to achieve parallelism, but they differ in how they utilize system resources and handle tasks.

***Multithreading*** involves running multiple threads within the same process, where all threads share the same memory space and resources. This makes communication between threads faster but also increases the risk of issues like race conditions and deadlocks if proper synchronization is not used. It is best suited for tasks that involve I/O operations or where tasks need to share data frequently, such as in web servers or GUI applications.

On the other hand, ***multiprocessing*** involves running multiple processes simultaneously, each with its own separate memory space. This eliminates the risk of memory-related conflicts and takes full advantage of multiple CPU cores, making it ideal for CPU-intensive tasks like data processing or complex calculations. However, since each process has its own memory, inter-process communication is slower and more complex than thread communication.

Hence, multithreading focuses on efficient use of shared resources within a single process, while multiprocessing aims for true parallel execution by using multiple independent processes.

Question 2: What are the challenges associated with memory management in Python?

-> Memory management in Python, while largely automated through its built-in garbage collector and reference counting mechanisms, still presents several challenges for developers. One key issue is memory leaks, which occur when objects are no longer needed but are still referenced somewhere in the program, preventing the garbage collector from reclaiming that memory. Another challenge is circular references, where two or more objects reference each other, making it difficult for the garbage collector to determine whether they can be safely deleted.

Additionally, Python’s Global Interpreter Lock (GIL) restricts true parallel execution of threads, which can limit efficient memory utilization in multi-threaded applications. Memory fragmentation is another problem, especially in long-running programs, where frequent allocation and deallocation of memory blocks can lead to inefficient use of available memory. Finally, inefficient handling of large data structures like lists or dictionaries, and the creation of unnecessary temporary objects, can cause excessive memory consumption. Developers often need to use tools like gc module, memory profilers, and optimized data structures (like generators or NumPy arrays) to manage memory effectively in Python applications.

```python
#Question 3:Write a Python program that logs an error message to a log file when a division by zero exception occurs.
import logging

# Configure logging settings
logging.basicConfig(
    filename='error_log.txt',       # Log file name
    level=logging.ERROR,            # Log only error-level messages
    format='%(asctime)s - %(levelname)s - %(message)s'  # Log format
)

try:
    # Example division operation
    numerator = int(input("Enter numerator: "))
    denominator = int(input("Enter denominator: "))
    result = numerator / denominator
    print(f"Result: {result}")

except ZeroDivisionError as e:
    logging.error("Division by zero occurred! Exception details: %s", e)
    print("Error: Division by zero is not allowed. Check the log file for details.")
```

```python
#Question 4:Write a Python program that reads from one file and writes its content to another file.
# Program to read from one file and write to another

# Specify file names
source_file = "source.txt"
destination_file = "destination.txt"

try:
    # Open the source file in read mode
    with open(source_file, 'r') as src:
        # Read all contents from the source file
        content = src.read()

    # Open the destination file in write mode
    with open(destination_file, 'w') as dest:
        # Write the content into the destination file
        dest.write(content)

    print(f"Content successfully copied from '{source_file}' to '{destination_file}'.")

except FileNotFoundError:
    print(f"Error: The file '{source_file}' does not exist.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
#Question 5: Write a program that handles both IndexError and KeyError using a try-except block.
# Program to handle IndexError and KeyError

try:
    # Example list and dictionary
    numbers = [10, 20, 30]
    student = {"name": "Nikhil", "age": 25}

    # Intentionally cause errors
    print(numbers[5])           # This will raise IndexError
    print(student["address"])   # This will raise KeyError

except IndexError:
    print("Error: You tried to access an index that does not exist in the list.")

except KeyError:
    print("Error: You tried to access a key that does not exist in the dictionary.")

except Exception as e:
    print(f"An unexpected error occurred: {e}")

finally:
    print("Program execution completed.")
```

Question 6: What are the differences between NumPy arrays and Python lists?

-> NumPy arrays and Python lists are both used to store collections of data, but they differ significantly in structure, performance, and functionality.

A Python list is a built-in, general-purpose data structure that can store elements of different data types (e.g., integers, strings, floats) in the same list. Lists are flexible but slower for numerical computations because operations must be performed element by element using loops. Memory consumption is also higher since each element in a list is a separate Python object.

In contrast, a NumPy array (from the NumPy library) is a powerful, specialized data structure designed for numerical and scientific computing. All elements in a NumPy array must be of the same data type, which allows for efficient memory usage and faster execution. NumPy supports vectorized operations, meaning mathematical operations can be applied to entire arrays at once without explicit loops. Additionally, NumPy provides advanced features like multi-dimensional arrays, broadcasting, and integration with other scientific libraries.

Hence, Python lists are more flexible and easy to use for general tasks, while NumPy arrays are optimized for performance and mathematical operations on large datasets.

Question 7:Explain the difference between apply() and map() in Pandas.

-> In Pandas, both apply() and map() are used to apply functions to data, but they differ in their scope and usage.

map() is primarily used on Series objects to apply a function element-wise. It can take a Python function, a dictionary, or a Series and transform each element in the original Series accordingly. For example, it is ideal for simple element-wise transformations, like multiplying numbers or replacing values in a single column.

apply(), on the other hand, is more flexible and can be used on both Series and DataFrame objects. When used on a Series, it behaves similarly to map(), but it can handle more complex operations that may return scalars or even new Series. When used on a DataFrame, apply() can operate row-wise or column-wise using the axis parameter, allowing you to perform computations that involve multiple columns or aggregate values.

In short:

map() → element-wise operation on a Series, simpler and limited.

apply() → element-wise on Series or row/column-wise on DataFrame, more versatile and suitable for complex operations.

```python
#Question 10: Calculate the correlation matrix using Seaborn and visualize it with a heatmap.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example dataset
data = {
    'Math': [85, 78, 92, 70, 88],
    'Physics': [80, 75, 90, 65, 85],
    'Chemistry': [78, 72, 88, 60, 80],
    'Biology': [82, 70, 85, 68, 90]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Calculate the correlation matrix
corr_matrix = df.corr()

# Display the correlation matrix
print("Correlation Matrix:\n", corr_matrix)

# Visualize the correlation matrix with a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
