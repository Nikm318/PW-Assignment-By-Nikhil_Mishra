Question 2: What are the key assumptions of Simple Linear Regression?

Answer:-

Simple Linear Regression (SLR) relies on several fundamental assumptions to ensure that the modelâ€™s estimates, interpretations, and predictions are valid and reliable.

Key Assumptions of Simple Linear Regression

Linearity

There is a linear relationship between the independent variable (X) and the dependent variable (Y).

This means the change in Y is proportional to the change in X.

If the relationship is not linear, SLR is not appropriate.

Independence of Errors

The residuals (errors) are independent of each other.

An error for one observation should not influence another.

This assumption is especially important for time-series data.

Homoscedasticity (Constant Variance)

The variance of the errors is constant across all values of X.

Residuals should have equal spread at all levels of the independent variable.

If variance changes (heteroscedasticity), predictions become less reliable.

Normality of Errors

The residuals are normally distributed with a mean of zero.

This assumption is mainly required for valid hypothesis testing and confidence intervals.

No Perfect Multicollinearity

Since Simple Linear Regression has only one predictor, this assumption is automatically satisfied.

(It becomes relevant in multiple linear regression.)

No Significant Outliers or Influential Points

There should be no extreme outliers that disproportionately affect the regression line.

Outliers can distort slope and intercept estimates.

Question 6: What is Logistic Regression? How does it differ from Linear Regression?

Answer:-

Logistic Regression is a supervised learning algorithm used for classification problems, where the dependent variable is categorical, most commonly binary (e.g., Yes/No, 0/1). Instead of predicting a continuous value, logistic regression estimates the probability that a given input belongs to a particular class using the sigmoid (logistic) function, which maps any real-valued input into a range between 0 and 1. The model then applies a threshold (typically 0.5) to convert this probability into a class label.

Logistic Regression differs from Linear Regression primarily in the type of output and modeling approach. Linear regression predicts continuous numerical values and assumes a linear relationship between the independent and dependent variables, whereas logistic regression predicts probabilities and models a non-linear relationship between the predictors and the probability of class membership. Linear regression uses the least squares method for estimation and can produce values outside the 0â€“1 range, making it unsuitable for classification tasks. In contrast, logistic regression uses maximum likelihood estimation, ensures outputs remain between 0 and 1, and is specifically designed for decision-making and classification scenarios such as spam detection, disease diagnosis, and credit risk assessment.

Question 7: Name and briefly describe three common evaluation metrics for regression models.

Answer:-

Three commonly used evaluation metrics for regression models are:

Mean Absolute Error (MAE)
MAE measures the average absolute difference between the actual and predicted values. It provides a clear interpretation of error in the same units as the target variable and treats all errors equally without heavily penalizing large deviations.

Mean Squared Error (MSE)
MSE calculates the average of the squared differences between actual and predicted values. By squaring the errors, it gives more weight to larger errors, making it useful when large prediction mistakes are particularly undesirable.

R-squared (RÂ²)
R-squared represents the proportion of variance in the dependent variable that is explained by the regression model. It indicates how well the model fits the data, with values closer to 1 signifying better explanatory power.

These metrics together help assess accuracy, error magnitude, and overall model fit in regression analysis.

Question 8: What is the purpose of the R-squared metric in regression analysis?

Answer:-  

The purpose of the R-squared (RÂ²) metric in regression analysis is to measure how well a regression model explains the variability of the dependent variable using the independent variable(s).

R-squared represents the proportion of total variation in the outcome variable that is explained by the model. Its value ranges from 0 to 1, where a higher value indicates a better model fit. For example, an R-squared value of 0.80 means that 80% of the variation in the dependent variable is explained by the regression model, while the remaining 20% is due to factors not captured by the model or random error.

In practical terms, R-squared helps analysts evaluate model effectiveness, compare different regression models, and understand the explanatory power of predictors. However, it does not indicate causation or guarantee good predictions, especially when overfitting is present, so it should be interpreted alongside other evaluation metrics.
